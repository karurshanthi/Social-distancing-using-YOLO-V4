```python
import cv2
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
cascPath = "haarcascade_pedestrian.xml"
```

```python
pedsCascade = cv2.CascadeClassifier(cascPath)
```

```python
image = cv2.imread('im3.jpg')
```

```python
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image)
plt.show()
```

```python
peds = pedsCascade.detectMultiScale(gray, scaleFactor=1.05, minNeighbors=5, minSize=(30,30))
```

```python
print("Found {0} pedestrians".format(len(peds)))
```

```python
for (x,y,w,h) in peds:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0,255,0), 2)
```

```python
plt.imshow(image)
```

```python
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
status = cv2.imwrite('res5.jpg', image)
print("image written to file-system : ", status)
```
